```python
IMG_DATA_PATH = "/home/eager/data/IXI/dti/"
ALL_IMG_PATHS = glob.glob(f"{IMG_DATA_PATH}/*.nii.gz")
```

```python
all_shapes = defaultdict(int)

for fpath in ALL_IMG_PATHS:
    im = nim.load_img(fpath)
    imdata = im.get_fdata()
    all_shapes[imdata.shape] += 1
```

```python
OUTPUT_DIR = "/home/eager/modeling-parkinsons/data/processed/ixi_dti_reshaped_control/"

for fpath in ALL_IMG_PATHS:
    im = nim.load_img(fpath)
    im_resample = nim.resample_img(im, im.affine, (116, 116, 72))
    
    basename = os.path.basename(fpath)
    fname = basename[:-7]
    
    info = fname.split("-")
    hospital = info[1]
    pid = info[2]
    idx = info[-1]
    iid = pid + "0000" + idx
    
    ofpath = f"{OUTPUT_DIR}/IXI_{pid}_{idx}_{hospital}_I{iid}.nii"
    
    im_resample.to_filename(ofpath)
```
